Outlier using IQR method 

```python
df.columns
```

```python
Q1=df.height.quantile(0.25)
Q3=df.height.quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
Upper_limit=Q3+1.5*IQR
print(lower_limit)
print(Upper_limit)
```

```python
Out=df[(df.height<lower_limit)| (df.height>Upper_limit)]
Out
```

```python
Out.size
```

```python
no_out=df[(df.height>lower_limit)& (df.height<Upper_limit)]
no_out
```
